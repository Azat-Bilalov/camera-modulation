### Последовательность работы при создании модели сцены и источника (Обьяснение)
Так как на вход мы получаем данные из спектрафотометра, которые имеют следующие количественные характеристики:

-Интенсивность излучения - длина волны по оси X и мощность излучения по оси Y

-Cпектральный коэффициент отражения

Таким образом мы можем сформировать спектральную ось, которая состоит из диапазона общих длин волн по оси X, и модели источника света, которая строиться на базе мощности излучения.

Данные значения удобны для перемножения с коэффициентом отражения объекта и дальнейшим расчетом.

Дальше создаётся модель объекта из коэффиентов отражения.

На базе полученных данных мы можем сформировать спектральную карту сцены, которая демонстрирует сколько света получиться отразить от обьекта, которые фиксирует объектив

_______________________________________________________________
В ходе улучшений были введены типизированные модели данных на основе *dataclass*. Спектральная ось была выделена в отдельную сущность *SpectralAxis*, которая хранит не только список длин волн, но и граничные значения диапазона и количество каналов. 

Это устраняет необходимость каждый раз вычислять эти параметры вручную или передавать их отдельными переменными(Для дальнейшей отладки параметры остаются в одном файле, и пока не берутся из внешних факторов).

Источник излучения и объект наблюдения были разделены в самостоятельные структуры *SourceConfig* и *ObjectConfig*, что отражает физическую природу задачи: источник и объект — это разные сущности с разными ролями в формировании сигнала сцены.


Функция *build_optic_input* была переработана так, чтобы принимать на вход уже построенные модели, а не сырые списки. Результат она теперь возвращает в виде объекта *SceneSignal*, а не словаря, что делает структуру выходных данных явной и проверяемой для следующей части разработки камеры. Дополнительно была введена вспомогательная функция *build_axis*, которая инкапсулирует построение спектральной оси из входного списка длин волн.
